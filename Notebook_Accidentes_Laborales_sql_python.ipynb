```python
import plotly.express as px
```

```python
accidentes_mes = df.groupby(df["FECHA DE OCURRENCIA DEL ACCIDENTE"].dt.to_period("M")).size().reset_index(name="Cantidad")

accidentes_mes["FECHA DE OCURRENCIA DEL ACCIDENTE"] = accidentes_mes["FECHA DE OCURRENCIA DEL ACCIDENTE"].astype(str)

fig = px.line(
    accidentes_mes,
    x="FECHA DE OCURRENCIA DEL ACCIDENTE",
    y="Cantidad",
    title="Accidentes por mes"
)

fig.show()
```

```python
conteo = df["SEDE"].value_counts().reset_index()

conteo.columns = ["SEDE","Cantidad"]

fig = px.bar(
    conteo,
    x="SEDE",
    y="Cantidad",
    title="Cantidad de accidentes por sede"
)

fig.show()
```

```python
conteo = df["TIPO DE ACCIDENTE "].value_counts().reset_index()

conteo.columns = ["Tipo","Cantidad"]

fig = px.pie(
    conteo,
    names="Tipo",
    values="Cantidad",
    title="Distribución por tipo de accidente"
)

fig.show()
```

```python
fig = px.histogram(
    df,
    x="HORAS MENSUAL",
    title="Distribución de horas mensuales"
)

fig.show()
```

```python
fig = px.box(
    df,
    y="FECHA DE OCURRENCIA DEL ACCIDENTE",
    title="Valores atípicos de fecha de ocurrencia de accidentes"
)

fig.show()
```

```python
import duckdb
```

```python
duckdb.sql("SELECT * FROM df LIMIT 5").df()
```

```python
duckdb.sql("""
SELECT *
FROM df
LIMIT 10
""").df()
```

```python
duckdb.sql("""
SELECT
    SEDE,
    COUNT(*) AS Cantidad_Accidentes
FROM df
GROUP BY SEDE
ORDER BY Cantidad_Accidentes DESC
""").df()
```

```python
duckdb.sql("""
SELECT
    "TIPO DE ACCIDENTE ",
    COUNT(*) AS Cantidad
FROM df
GROUP BY "TIPO DE ACCIDENTE "
ORDER BY Cantidad DESC
""").df()
```

```python
duckdb.sql("""
SELECT
AVG("HORAS MENSUAL") AS Promedio_Horas
FROM df
""").df()
```

```python
duckdb.sql("""
SELECT
YEAR("FECHA DE OCURRENCIA DEL ACCIDENTE") AS Año,
COUNT(*) AS Accidentes
FROM df
GROUP BY Año
ORDER BY Año
""").df()
```
